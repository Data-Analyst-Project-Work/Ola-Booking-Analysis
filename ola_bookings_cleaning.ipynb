### Import Libraries

```python
import pandas as pd
```

### Load Dataset

```python
ola_df = pd.read_excel("dataset/ola_bookings.xlsx")
ola_df.head()
```

### Dataset Information

```python
ola_df.info()
```

### Replace NaN Values

```python
ola_df['V_TAT'] = ola_df['V_TAT'].fillna(0)
ola_df['C_TAT'] = ola_df['C_TAT'].fillna(0)
ola_df['Canceled_Rides_by_Customer'] = ola_df['Canceled_Rides_by_Customer'].fillna("No")
ola_df['Canceled_Rides_by_Driver'] = ola_df['Canceled_Rides_by_Driver'].fillna("No")
ola_df['Incomplete_Rides'] = ola_df['Incomplete_Rides'].fillna("Yes")
ola_df['Incomplete_Rides_Reason'] = ola_df['Incomplete_Rides_Reason'].fillna("No")
ola_df['Payment_Method'] = ola_df['Payment_Method'].fillna("No")
ola_df['Driver_Ratings'] = ola_df['Driver_Ratings'].fillna(0.0)
ola_df['Customer_Rating'] = ola_df['Customer_Rating'].fillna(0.0)
```

### Checking NULL Values

```python
ola_df.isnull().sum()
```

### Show Cleaned Dataset

```python
ola_df.sample(10)
```

```python
ola_df['Date'] = pd.to_datetime(ola_df['Date'])
ola_df['Date'] = ola_df['Date'].dt.strftime('%d-%m-%Y')

ola_df.head()
```

### Export to CSV

```python
ola_df.to_csv('dataset/ola_bookings_cleaned.csv', index=False)
```
