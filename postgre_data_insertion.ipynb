### Import Libraries

```python
import pandas as pd
import psycopg2
```

### Load Cleaned Dataset

```python
ola_df = pd.read_csv("dataset/ola_bookings_cleaned.csv")
ola_df.sample(10)
```

### Load PostgreSQL Connection, Create & Insert Data

```python
csv_file_path = 'dataset/ola_bookings_cleaned.csv'
db_config = {
    'dbname': 'OlaDB',
    'user': 'postgres',
    'password': 'postgres',
    'host': 'localhost',
    'port': 5432
}

table_name = csv_file_path.split('/')[-1].split('.')[0]

df = pd.read_csv(csv_file_path)

columns = df.columns
column_definitions = ', '.join([f'"{col}" VARCHAR' for col in columns])  # All columns as VARCHAR
create_table_query = f'CREATE TABLE IF NOT EXISTS "{table_name}" ({column_definitions});'

conn = None

try:
    conn = psycopg2.connect(**db_config)
    cursor = conn.cursor()
    
    cursor.execute(create_table_query)
    conn.commit()
    print(f"Table '{table_name}' created successfully.")
    
    for _, row in df.iterrows():
        values = "', '".join([str(val).replace("'", "''") for val in row])
        column_names = ', '.join([f'"{col}"' for col in columns])
        insert_query = f"INSERT INTO \"{table_name}\" ({column_names}) VALUES ('{values}');"
        cursor.execute(insert_query)
    conn.commit()
    print(f"Data inserted successfully into '{table_name}'.")
    
except Exception as e:
    print("Error:", e)

finally:
    if conn:
        cursor.close()
        conn.close()
        print("Database connection closed.")
```
